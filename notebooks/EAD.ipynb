# 1. Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.float_format', lambda x: '%.2f' % x)

sns.set_style('whitegrid')
```

# 2. Load Data

```python
df = pd.read_parquet('../data/raw/imovirtual_v2.parquet')
print(df.shape)
df.head()
```

```python
df['type_ads'] = df.company.apply(lambda x: 'Particular' if x == 'Anúncio Particular' else 'Corretora')
```

```python
df = df.loc[df.price != "Preço sob consulta"]
df = df.reset_index(drop=True)
```

```python
df['metric'] = df['metric'].str.replace(',', '.')
df.metric = df.metric.str.replace(' m²', '').str.replace(' ', '').astype(float)
df.rooms = df.rooms.str.replace('T', '').str.replace(' ou superior','').astype(int)
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
df['price'] = df['price'].str.replace(' ', '').astype(float)
```

```python
df.describe(percentiles=[0.02, 0.1, .25, .50, 0.9, 0.95, 0.99, 0.999])
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
df.bathroom.fillna(1, inplace=True)
```

```python
min_price = np.percentile(df.price, 0.02)
print(min_price)
max_price = np.percentile(df.price, 99)
print(max_price)
```

```python
df = df.query('price >= @min_price and price <= @max_price')
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
min_metric = np.percentile(df.metric, 0.02)
max_metric = np.percentile(df.metric, 99.9)
print(f'Min_metric: {min_metric}, Max_metric: {max_metric}')

df = df.query('metric >= @min_metric and metric <= @max_metric')
df.reset_index(drop=True, inplace=True)
```

```python
min_bathroom = np.percentile(df.bathroom, 0.02)
max_bathroom = np.percentile(df.bathroom, 99)
print(f'min_bathroom: {min_bathroom}, max_bathroom: {max_bathroom}')

```

```python
df = df.query('bathroom >= @min_bathroom and bathroom <= @max_bathroom')
df.reset_index(drop=True, inplace=True)
print(df.shape)
```

```python
min_rooms = np.percentile(df.rooms, 0.02)
max_rooms = np.percentile(df.rooms, 99)
print(f'min_rooms: {min_rooms}, max_rooms: {max_rooms}')

```

```python
df = df.query('rooms >= @min_rooms and rooms <= @max_rooms')
df.reset_index(drop=True, inplace=True)
print(df.shape)
```
